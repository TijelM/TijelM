**Object Oriented method**

```python
import matplotlib.pyplot as plt
%matplotlib inline

import numpy as np

x = np.linspace(0,5,11)
y = x**2

fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])  #left, bottom, width and height

axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes2.plot(y,x)
```

```python
fig,axes = plt.subplots(nrows = 3, ncols = 3)

plt.tight_layout() # removes overlaps between plots
```

```python
fig,axes = plt.subplots(nrows = 1, ncols = 2)

axes[0].plot(x,y) #axes is a matplotlib array through which a user can iterate
axes[0].set_title('First plot')

axes[1].plot(y,x)
axes[1].set_title('Second title')
```

```python
# Adding legends to a plot

fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x**2,label = 'X Squared')
ax.plot(x,x**3,label = 'X Cubed')

ax.legend()  #location of the legend can be specified by ax.legend(loc=0) or any other value till 10. 0 is the best
```

```python
#customizing the plot

fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color = 'red',linewidth=3,alpha=0.5,linestyle='-',marker='o',markersize=10,
       markerfacecolor='blue',markeredgewidth=3,markeredgecolor='green')




#default linewidth is 1, we can also write 'lw'; alpha is the transparency of the line; ;linestyle or ls examples = --,-.,steps,: etc.
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,y,color = 'red',lw=2,ls='-')

ax.set_xlim([0,1])  #for setting limits to the plot
ax.set_ylim([0,2])
```
